```python
Decision: Which pages should the content team review first? Action: A human editor will look at the flagged page and decide whether to update the content, rewrite the title, or merge it. Cost of a wrong call: If the model flags the wrong page (False Positive), a human wastes 5 minutes reviewing a page that didn't need help. If the model misses a declining page (False Negative), the client loses traffic they could have saved.
```
